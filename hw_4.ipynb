Решите аналитически и потом численно (в программе) уравнение, зависящее от параметра а: 
sin(а*x)=0
при условии: 0.01<a<0.02, 100<х<500.
Т.е. надо найти решение х как функцию параметра а - построить график x=x(а).
Если численным методом не получается найти все ветви решения x(а), то отыщите хотя бы одну.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

a = np.linspace(np.pi / 500, np.pi / 100, 1000)
x2 = 3 * np.pi / a
x1 =  2 * np.pi / a
x = np.pi / a
plt.plot(a,x, color='blue')
plt.plot(a,x1,color='red')
plt.plot(a,x2, color='green')
plt.xlabel('a')
plt.ylabel('x')
plt.xlim(0.01,0.02)
plt.ylim(100,500)
plt.plot(0.01, 100 * np.pi, marker="o", color='blue')
plt.plot(0.02, 50 * np.pi, marker="o", color='blue')
plt.plot(np.pi / 250, 500, marker="o", color='red')
plt.plot(0.02, 100* np.pi, marker="o", color='red')
plt.plot(3*np.pi/500, 500, marker="o", color='green')
plt.plot(0.02, 150*np.pi, marker="o", color='green')

plt.grid(True)
plt.show()
```
